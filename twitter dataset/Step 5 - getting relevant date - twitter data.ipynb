# Getting relevant date

```python
import pandas as pd
df = pd.read_csv('twitter_dropdup_trollremoved_noninfluencers_withemotioncat.csv')
```

```python
df.columns
```

```python
df['Date'] = df['Date'].astype('datetime64[ns]')
```

```python
df_enddec = df.loc[df['Date'] < "2022-01-01", :]
```

```python
df_enddec.tail()
```

```python
len(df_enddec)
```

```python
df_enddec.to_csv('twitter_data_janonwards.csv')
```

```python
df_enddec_fromjuly = df_enddec.loc[df_enddec['Date'] > "2020-06-30", :]
```
