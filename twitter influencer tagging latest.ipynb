```python
import pandas as pd

df1= pd.read_csv('SG_twitter_trollremoved_V3_26AUG.csv')
```

```python
df1.head()
```

```python
df1['influencer_score']=df1['user_followers']/df1['user_friends']
```

```python
df1.head()
```

```python
def det_info_status(row):
    if float((row['influencer_score'])) > 1:
        val = "info-disseminators"
    else:
        val = "info-seeking"
    return val
```

```python
def det_info_status(row):
    if float((row['influencer_score'])) > 1:
        val = 0
    else:
        val = 1
    return val
```

```python
df1["info_status"] = df1.apply(det_info_status, axis=1)
```
