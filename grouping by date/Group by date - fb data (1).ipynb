```python
import nltk
import datetime as datetime
import pandas as pd
pd.set_option('display.max_rows', None)
    
df = pd.read_csv('fb data_all_processed.csv')
```

```python
df.columns
```

```python
df.head()
```

```python
#use agg 'count' for all non-intensity score vales
total_count = df.groupby(['date']).agg(['count'])
total_count.tail(30)
```

```python
total_count.to_csv('fb_total_count.csv')
```
