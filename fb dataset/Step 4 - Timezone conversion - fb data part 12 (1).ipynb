```python
import pandas as pd
new_df = pd.read_csv('vaccine_fb_sgonly_droppeddup_trollremoved.csv')
```

```python
new_df.columns
```

```python
new_df.head()
```

```python
len(new_df)
```

```python
new_df['utc_time'] = pd.to_datetime(new_df['creation_time'], unit='s', errors='coerce')
```

```python
new_df['sg_time']= new_df['utc_time'] + pd.DateOffset(hours=16)
```

```python
new_df['date'] = new_df['sg_time'].dt.date
new_df.head()
```

```python
new_df = new_df.sort_values(by="sg_time", ascending=True)
```

```python
new_df.head()
```

```python
startdate = pd.to_datetime("2020-06-30").date()
new_df1 = new_df.loc[new_df['date'] > startdate, :]
new_df1.sort_values(by='date', inplace = True, ascending=True)
new_df1.head()
```

```python
enddate = pd.to_datetime("2022-01-01").date()
new_df1 = new_df1.loc[new_df1['date'] < enddate, :]
new_df1.sort_values(by='date', inplace = True, ascending=True)
new_df1.tail()
```

```python
len(new_df1)
```

```python
new_df1.to_csv('fb_part12_sgonly_droppeddup_trollremoved_reldate.csv')
```
