```python
import pandas as pd
new_df = pd.read_csv('fb comments omicron only.csv')
```

```python
new_df.columns
```

```python
new_df.head()
```

```python
new_df['utc_time'] = pd.to_datetime(new_df['creation_time'], unit='s', errors='coerce')
```

```python
new_df['sg_time']= new_df['utc_time'] + pd.DateOffset(hours=16)
```

```python
new_df['date'] = new_df['sg_time'].dt.date
new_df.head()
```

```python
new_df = new_df.sort_values(by="sg_time", ascending=True)
```

```python
new_df.tail()
```

```python
enddate = pd.to_datetime("2022-01-01").date()
new_df = new_df.loc[new_df['date'] < enddate, :]
new_df.sort_values(by='date', inplace = True, ascending=True)
new_df.tail()
```

```python
new_df.to_csv('fb comments omicron only with correct date.csv')
```
